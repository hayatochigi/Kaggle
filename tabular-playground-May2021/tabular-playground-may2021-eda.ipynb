```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

# Read/Concatenate train and test files.

```python
train = pd.read_csv('../input/tabular-playground-series-may-2021/train.csv')
test = pd.read_csv('../input/tabular-playground-series-may-2021/test.csv')
dataset = pd.concat([train, test], axis=0)
dataset.head()
```

# Visualize feature_x

```python
_ = plt.figure(figsize=(20,30))
for i in range(0,len(dataset.columns)-2):
    _ = plt.subplot(10,5,i+1)
    col = 'feature_' + str(i)
    _ = plt.hist(dataset[col], bins=20)
    _ = plt.title(col)

_ = plt.tight_layout()
```

# continuous or categorical?

```python
for x in dataset.columns:
    unique_size = len(dataset[x].unique())
    print(f'unique {x}:\t{unique_size}')
```
