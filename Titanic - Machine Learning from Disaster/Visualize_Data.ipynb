```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
plt.style.use('ggplot')
```

```python
dataset = pd.read_csv('data/train.csv')
visualize_data = dataset[['Embarked', 'Survived', 'PassengerId']].dropna().groupby(['Embarked', 'Survived']).count().unstack()
visualize_data.plot.bar(stacked=True)

# Survived Ratiodataframe
visualize_data['Survived_ratio'] = visualize_data.iloc[:,0] / (visualize_data.iloc[:,0] + visualize_data.iloc[:,1])
print(visualize_data)
```

```python
histogram_data = dataset[['Survived', 'Age']]
hist_died = histogram_data.loc[histogram_data['Survived'] == 0, 'Age'].dropna().values
hist_alive = histogram_data.loc[histogram_data['Survived'] == 1, 'Age'].dropna().values

plt.hist((hist_died,hist_alive), histtype='barstacked', bins=10, label=('Death', 'Survive'))
plt.legend()
```

```python
women = dataset[(dataset['Sex'] == 'female')]
women_survived = women[women['Survived'] == 1]
print(f'% of women who survived: {len(women_survived)/len(women) * 100: .2f}%')

men = dataset[(dataset['Sex'] == 'male')]
men_survived = men[men['Survived'] == 1]
print(f'% of men who survived: {len(men_survived)/len(men) * 100: .2f}%')
```
